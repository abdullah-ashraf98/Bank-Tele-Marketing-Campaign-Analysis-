```python
filtrd1=bank.groupby('job')['deposit'].agg(lambda x:x.mode().loc[0])
filtrd1
```

```python
deposit_counts = bank['deposit'].value_counts()
deposit_percent = bank['deposit'].value_counts(normalize=True) * 100
print("Deposit distribution counts", deposit_counts)
print("Deposit distribution (percentage", deposit_percent.round(2))

```

```python
filtered = bank[bank['deposit'] == 'yes']
yes = filtered.groupby(['job', 'marital']).size().unstack().fillna(0).astype(int)
yes.corr()
```

```python
yes.corr()
```

```python
pivot_age = bank.pivot_table(index='deposit', values='age', aggfunc=['mean', 'median', 'min', 'max'])
print(pivot_age)

```

```python
# Job vs deposit
pivot_job = bank.pivot_table(index='job', columns='deposit', aggfunc='size', fill_value=0)
print(pivot_job)
```

```python
pivot_marital = bank.pivot_table(index='marital', columns='deposit', aggfunc='size', fill_value=0)
print(pivot_marital)
```

```python
pivot_education = bank.pivot_table(index='education', columns='deposit', aggfunc='size', fill_value=0)
print(pivot_education)

```

```python
pivot_duration = bank.pivot_table(index='deposit', values='duration', aggfunc=['mean', 'median', 'min', 'max'])
print(pivot_duration)
```

```python
pivot_month = bank.pivot_table(index='month', columns='deposit', aggfunc='size', fill_value=0)
print(pivot_month)
```

```python
pivot_day = bank.pivot_table(index='day_of_week', columns='deposit', aggfunc='size', fill_value=0)
print(pivot_day)
```

```python
economic_vars = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
for var in economic_vars:
    pivot_econ = bank.pivot_table(index='deposit', values=var, aggfunc=['mean', 'median', 'min', 'max'])
    print(pivot_econ)

```

```python
loan_vars = ['credit', 'housing loan', 'loan']
for var in loan_vars:
    loan_deposit = bank.groupby([var, 'deposit']).size()
    print( loan_deposit)
```

```python
campaign_stats = bank.groupby('deposit')['campaign'].describe()
previous_stats = bank.groupby('deposit')['previous'].describe()
print( "current campaign\n",  campaign_stats)
print('pre campaign\n' , previous_stats)
```
