```python
def outlier(columns):
   q1=columns.quantile(.25)
   q3=columns.quantile(.75)
   iqr=q3-q1
   lower = q1- 1.5*iqr
   upper= q3+1.5*iqr
   return [float(iqr),float(lower),float(upper)]
print(outlier(bank['age']))

```
